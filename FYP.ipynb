# One Hot Encoding
Another widely used method for handling categorical values is one-hot encoding. It merely generates new features based on how many distinct values there are in the category feature. The category's individual values will all be introduced as features.

Since, our dataset does not demand a degree of importance or a weightage to the categorical variables and they are simply distinct to each other, we perform One Hot Encoding

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
def train1(model, X_train, Y1_train,X_test,Y1_test):
    # training the model
    model.fit(X_train, Y1_train)
    
    pred = model.predict(X_test)
    # perform cross-validation
    cv_score = cross_val_score(model, X_train, Y1_train, scoring='neg_mean_squared_error', cv=15)
    cv_score = np.abs(np.mean(cv_score))
    
    print("Model Report")
    print("CV Score:", cv_score)
    print("R2_Score:", r2_score(Y1_test,pred))

def train2(model, X_train, Y2_train, X_test,Y2_test):
    # training the model
    model.fit(X_train, Y2_train)
    
    pred = model.predict(X_test)
    # perform cross-validation
    cv_score = cross_val_score(model, X_test, Y2_test, scoring='neg_mean_squared_error', cv=15)
    cv_score = np.abs(np.mean(cv_score))
    
    print("Model Report")
    print("CV Score:", cv_score)
    print("R2_Score:", r2_score(Y2_test,pred))
    
def train3(model, X_train, Y3_train,X_test,Y3_test):
    # training the model
    model.fit(X_train, Y3_train)
    
    pred = model.predict(X_test)
    # perform cross-validation
    cv_score = cross_val_score(model, X_train, Y3_train, scoring='neg_mean_squared_error', cv=15)
    cv_score = np.abs(np.mean(cv_score))
    
    print("Model Report")
    print("CV Score:", cv_score)
    print("R2_Score:", r2_score(Y3_test,pred))
```

Training models train1, train2, train3 has been defined for predicting roughness, tensile strength and elongation respectively.

```python
import sklearn
print(sklearn.__version__)
```

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor(random_state=0)
train1(model,x_train, y1_train,x_test,y1_test)
coef = pd.Series(model.feature_importances_, x.columns).sort_values(ascending=False)
path= model.cost_complexity_pruning_path(x_train, y1_train)
ccp_alphas,impurities = path.ccp_alphas, path.impurities
trees=[]
for ccp_alpha in ccp_alphas:
    tree = DecisionTreeRegressor(random_state=0, ccp_alpha=ccp_alpha)
    tree.fit(x_train, y1_train)
    trees.append(tree)
mae = []
for tree in trees:
    y1_pred = tree.predict(x_test)
    mae.append(np.mean(np.abs(y1_test - y1_pred)))
fig, ax = plt.subplots()
ax.plot(ccp_alphas[:-1], mae[:-1], marker='o', drawstyle="steps-post")
ax.set_xlabel("Effective alpha")
ax.set_ylabel("Mean absolute error")
ax.set_title("Mean absolute error vs effective alpha")

plt.show()
```
